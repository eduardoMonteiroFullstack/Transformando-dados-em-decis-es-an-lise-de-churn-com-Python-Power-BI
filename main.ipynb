```python
try:
    df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv", encoding="UTF-8")
    #print(df)
    df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
    df = df.dropna(subset=['TotalCharges'])
    df.columns = df.columns.str.strip()
    df.to_excel("planilha_tratada.xlsx", index=False)
except Exception as e:
    print(e)
```

# METODO DE PAGAMENTO VS DESISTÊNCIA

```python
print(df.groupby('PaymentMethod')['Churn'].value_counts(normalize=True))
```


# Taxa de desistencia avaliando por contratos


```python
print(df.groupby('Contract')['Churn'].value_counts(normalize=True) * 100)
```
